```python
import pandas as pd
import numpy as np
from google.colab import files
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.vector_ar.vecm import coint_johansen
from statsmodels.stats.diagnostic import acorr_ljungbox
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# Function to upload and load Excel files for a stock
def upload_stock_data(stock_name):
    print(f"Please upload {stock_name} 1-minute Excel file")
    upload = files.upload()
    filename = list(upload.keys())[0]
    df = pd.read_excel(filename)
    return df

# Upload Excel files for VISA, JNJ, NVIDIA, and LMT (1-minute data)
stocks = ['VISA', 'JNJ', 'NVIDIA', 'LMT']
stock_data = {}

for stock in stocks:
    stock_data[stock] = upload_stock_data(stock)

# Print available columns for each DataFrame
for stock in stocks:
    print(f"\n{stock} 1-minute DataFrame columns:", list(stock_data[stock].columns))

# Ask user for the column name containing closing prices
close_column = input("Please enter the column name containing closing prices: ")

# Extract and clean closing prices for each stock
closing_prices = {}
try:
    for stock in stocks:
        closing_prices[stock] = stock_data[stock][close_column]
        # Remove NaN and infinite values
        closing_prices[stock] = closing_prices[stock].replace([np.inf, -np.inf], np.nan).dropna()
        if closing_prices[stock].empty:
            raise ValueError(f"No valid data for {stock} after removing NaN/inf")
except KeyError:
    print(f"Error: Column '{close_column}' not found in one or more DataFrames")
    print("Please check the column names and try again")
    raise

# Ensure all series have the same length
min_length = min(len(closing_prices[stock]) for stock in stocks)
for stock in stocks:
    closing_prices[stock] = closing_prices[stock][:min_length]

# Stationarity Tests (ADF and KPSS)
def unit_root_tests(series, name, diff_count=0):
    # Ensure series has no NaN or infinite values
    series = series.replace([np.inf, -np.inf], np.nan).dropna()
    if series.empty:
        print(f"Error: {name} (Diff {diff_count}) is empty after cleaning")
        return False

    print(f"\nUnit Root Tests for {name} (Diff {diff_count}):")

    # ADF Test
    try:
        adf_result = adfuller(series)
        print("ADF Test:")
        print(f'ADF Statistic: {adf_result[0]:.4f}')
        print(f'p-value: {adf_result[1]:.4f}')
        print(f'Critical Values: {adf_result[4]}')
        print("Interpretation:")
        if adf_result[1] < 0.05:
            print(f"  - p-value < 0.05: Reject null hypothesis - {name} is stationary")
        else:
            print(f"  - p-value >= 0.05: Fail to reject null - {name} may be non-stationary")
    except:
        print(f"ADF test failed for {name} (Diff {diff_count})")
        return False

    # KPSS Test
    try:
        kpss_result = kpss(series)
        print("\nKPSS Test:")
        print(f'KPSS Statistic: {kpss_result[0]:.4f}')
        print(f'p-value: {kpss_result[1]:.4f}')
        print(f'Critical Values: {kpss_result[3]}')
        print("Interpretation:")
        if kpss_result[1] < 0.05:
            print(f"  - p-value < 0.05: Reject null hypothesis - {name} is non-stationary")
        else:
            print(f"  - p-value >= 0.05: Fail to reject null - {name} may be stationary")
    except:
        print(f"KPSS test failed for {name} (Diff {diff_count})")
        return False

    return adf_result[1] < 0.05 and kpss_result[1] >= 0.05

# Ensure stationarity by differencing if needed
def make_stationary(series, name, max_diff=2):
    diff_count = 0
    current_series = series.copy()

    while diff_count <= max_diff:
        if unit_root_tests(current_series, name, diff_count):
            print(f"\n{name} is stationary after {diff_count} differences")
            return current_series, diff_count
        current_series = current_series.diff().dropna()
        diff_count += 1
        if current_series.empty:
            print(f"Error: {name} became empty after {diff_count} differences")
            return series, diff_count

    print(f"\nWarning: {name} could not be made stationary after {max_diff} differences")
    return current_series, diff_count

# Apply stationarity tests and differencing
stationary_prices = {}
diff_counts = {}
for stock in stocks:
    stationary_series, diffs = make_stationary(closing_prices[stock], stock)
    stationary_prices[stock] = stationary_series
    diff_counts[stock] = diffs

# Cointegration Test (Johansen) on original (non-differenced) data
def cointegration_test(df):
    # Ensure no NaN or infinite values
    df = df.replace([np.inf, -np.inf], np.nan).dropna()
    if df.empty:
        print("Error: Cointegration data is empty after cleaning")
        return

    try:
        result = coint_johansen(df, det_order=0, k_ar_diff=1)
        print(f"\nJohansen Cointegration Test (1-minute):")
        print(f"Trace statistic: {result.lr1}")
        print(f"Critical values (90%, 95%, 99%): {result.cvt}")
        print("Interpretation:")
        for i in range(len(result.lr1)):
            if result.lr1[i] > result.cvt[i, 1]:
                print(f"  - r = {i}: Cointegration exists at 95% confidence level")
                print(f"    Trace statistic ({result.lr1[i]:.2f}) > 95% critical value ({result.cvt[i, 1]:.2f})")
            else:
                print(f"  - r = {i}: No cointegration at 95% confidence level")
                print(f"    Trace statistic ({result.lr1[i]:.2f}) <= 95% critical value ({result.cvt[i, 1]:.2f})")
        if result.lr1[0] > result.cvt[0, 1]:
            print(f"Conclusion: VISA, JNJ, NVIDIA, and LMT are cointegrated")
        else:
            print(f"Conclusion: No evidence of cointegration")
    except:
        print("Error: Johansen cointegration test failed")

# Prepare data for cointegration
coint_df = pd.DataFrame({
    'VISA': closing_prices['VISA'],
    'JNJ': closing_prices['JNJ'],
    'NVIDIA': closing_prices['NVIDIA'],
    'LMT': closing_prices['LMT']
}).dropna()
cointegration_test(coint_df)

# Function to fit and evaluate AR, ARMA, and ARIMA models on stationary data
def fit_time_series_models(series, name, diff_count, max_p=3, max_q=3):
    best_models = {'AR': None, 'ARMA': None, 'ARIMA': None}
    best_aics = {'AR': float('inf'), 'ARMA': float('inf'), 'ARIMA': float('inf')}
    best_orders = {'AR': None, 'ARMA': None, 'ARIMA': None}

    # AR model: (p, 0, 0) on stationary series
    for p in range(1, max_p + 1):
        try:
            model = ARIMA(series, order=(p, 0, 0))
            results = model.fit()
            if results.aic < best_aics['AR']:
                best_aics['AR'] = results.aic
                best_orders['AR'] = (p, 0, 0)
                best_models['AR'] = results
        except:
            continue

    # ARMA model: (p, 0, q) on stationary series
    for p in range(1, max_p + 1):
        for q in range(1, max_q + 1):
            try:
                model = ARIMA(series, order=(p, 0, q))
                results = model.fit()
                if results.aic < best_aics['ARMA']:
                    best_aics['ARMA'] = results.aic
                    best_orders['ARMA'] = (p, 0, q)
                    best_models['ARMA'] = results
            except:
                continue

    # ARIMA model: (p, d, q) on original series (using diff_count as d)
    for p in range(max_p + 1):
        for q in range(max_q + 1):
            try:
                model = ARIMA(closing_prices[name], order=(p, diff_count, q))
                results = model.fit()
                if results.aic < best_aics['ARIMA']:
                    best_aics['ARIMA'] = results.aic
                    best_orders['ARIMA'] = (p, diff_count, q)
                    best_models['ARIMA'] = results
            except:
                continue

    # Print model summaries
    print(f"\nModel Comparison for {name} (1-minute):")
    for model_type in ['AR', 'ARMA', 'ARIMA']:
        if best_orders[model_type]:
            print(f"\nBest {model_type} model:")
            print(f"Order: {best_orders[model_type]}")
            print(f"AIC: {best_aics[model_type]:.2f}")
            print("Interpretation:")
            p, d, q = best_orders[model_type]
            print(f"  - p={p}: {p} autoregressive term(s)")
            print(f"  - d={d}: {d} difference(s) applied")
            print(f"  - q={q}: {q} moving average term(s)")

            # Residual diagnostics
            residuals = best_models[model_type].resid
            lb_test = acorr_ljungbox(residuals, lags=[10], return_df=True)
            print("\nResidual Diagnostics:")
            print(f"Ljung-Box Test (lag 10): p-value = {lb_test['lb_pvalue'].iloc[0]:.4f}")
            if lb_test['lb_pvalue'].iloc[0] > 0.05:
                print("  - p-value > 0.05: Residuals appear to be white noise (good model fit)")
            else:
                print("  - p-value <= 0.05: Residuals show autocorrelation (model may need improvement)")

            # Plot residual ACF
            sm.graphics.tsa.plot_acf(residuals, lags=20, title=f'{name} (1-minute) {model_type} Residual ACF')
            plt.show()

    # Select best model based on AIC
    best_model_type = min(best_aics, key=best_aics.get)
    print(f"\nBest overall model for {name} (1-minute): {best_model_type} with AIC {best_aics[best_model_type]:.2f}")

    return best_models[best_model_type], best_orders[best_model_type]

# Fit AR, ARMA, and ARIMA models for each stock
best_models = {}
best_orders = {}
for stock in stocks:
    model, order = fit_time_series_models(stationary_prices[stock], stock, diff_counts[stock])
    best_models[stock] = model
    best_orders[stock] = order

# Forecast next 30 periods for each stock using best ARIMA model
forecast_steps = 30
forecasts = {}
for stock in stocks:
    # Re-fit ARIMA on original series with determined differencing
    model = ARIMA(closing_prices[stock], order=best_orders[stock])
    fitted_model = model.fit()
    forecasts[stock] = fitted_model.forecast(steps=forecast_steps)

# Create forecast index for plotting
historical_length = len(closing_prices[stocks[0]])
forecast_index = np.arange(historical_length, historical_length + forecast_steps)

# Plot original series with forecasts
plt.figure(figsize=(12,6))
for stock, color in [('VISA', 'red'), ('JNJ', 'blue'), ('NVIDIA', 'green'), ('LMT', 'purple')]:
    plt.plot(closing_prices[stock], label=f'{stock} Historical')
    plt.plot(forecast_index, forecasts[stock], label=f'{stock} Forecast', color=color)
plt.title('Stock Closing Prices with Forecasts (1-minute)')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()

# Debug print to verify forecast alignment
for stock in stocks:
    print(f"\nDebug for {stock}: Last historical index = {historical_length - 1}, Forecast starts at {forecast_index[0]}, First 5 forecast values = {forecasts[stock][:5].tolist()}")

# Detailed forecast plot with confidence intervals and interpretation
def plot_forecast(model, series, name, steps=30):
    forecast_obj = model.get_forecast(steps=steps)
    forecast = forecast_obj.predicted_mean
    conf_int = forecast_obj.conf_int()

    # Ensure forecast index starts after historical data
    forecast_index = np.arange(len(series), len(series) + steps)

    plt.figure(figsize=(12,6))
    plt.plot(series.index, series.values, label=f'{name} Historical')  # Use index if available
    plt.plot(forecast_index, forecast, label='Forecast', color='red')
    plt.fill_between(forecast_index,
                    conf_int.iloc[:, 0],
                    conf_int.iloc[:, 1],
                    color='pink',
                    alpha=0.3,
                    label='95% Confidence Interval')
    plt.title(f'{name} Price Forecast (1-minute)')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

    # Forecast interpretation
    last_value = series.iloc[-1]
    mean_forecast = forecast.mean()
    print(f"\nForecast Interpretation for {name} (1-minute):")
    print(f"Last observed value: {last_value:.2f}")
    print(f"Average forecast value: {mean_forecast:.2f}")
    print(f"Forecast change: {mean_forecast - last_value:.2f}")
    if mean_forecast > last_value:
        print("Trend: Upward forecast trend")
    elif mean_forecast < last_value:
        print("Trend: Downward forecast trend")
    else:
        print("Trend: Flat forecast trend")
    print(f"95% CI range at period {steps}: [{conf_int.iloc[-1, 0]:.2f}, {conf_int.iloc[-1, 1]:.2f}]")

# Generate detailed forecast plots and interpretations for each stock
for stock in stocks:
    model = ARIMA(closing_prices[stock], order=best_orders[stock])
    fitted_model = model.fit()
    plot_forecast(fitted_model, closing_prices[stock], stock)

# Print forecast values for the first 5 periods
for stock in stocks:
    print(f"\n{stock} Forecast Values (next 5 periods, 1-minute):")
    print(forecasts[stock][:5])
```
